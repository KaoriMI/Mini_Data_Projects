```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_red = pd.read_csv('winequality-red.csv', sep = ';')
df_white = pd.read_csv('winequality-white.csv', sep = ';')
```

```python
df_red.head()
```

```python
df_white.head()
```

```python
df_white.columns
```

```python
new_col = []
for x in df_white.columns:
    x = x.replace(' ', '_')
    new_col.append(x)
```

```python
new_col
```

```python
df_red.columns = new_col
df_white.columns = new_col
```

```python
df_red.head()
```

```python
color_red = np.repeat('red', df_red.shape[0])
color_white = np.repeat('white', df_white.shape[0])
```

```python
df_red["color"] = color_red
df_white["color"] = color_white
```

```python
df_red.head()
```

```python
df_wine = df_red.append(df_white)
```

```python
df_wine.head()
```

```python
df_wine.to_csv('wine.csv', index = False)
```
